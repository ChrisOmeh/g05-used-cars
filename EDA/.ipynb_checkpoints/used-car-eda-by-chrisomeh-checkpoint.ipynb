## Context

Craigslist is the world's largest collection of used vehicles for sale, yet it's very difficult to collect all of them in the same place. I built a scraper for a school project and expanded upon it later to create this dataset which includes every used vehicle entry within the United States on Craigslist.

# Content

This data is scraped every few months, it contains most all relevant information that Craigslist provides on car sales including columns like <b> price, condition, manufacturer, latitude/longitude, and 18 other categories.</b> For ML projects, consider feature engineering on location columns such as long/lat. For previous listings, check older versions of the dataset.

* Craiglist being world leading company dealing on used cars, there is massive information about used cars in craiglist website and databases. So, how can we analyze this dataset to get useful business insights about used cars in USA? Can a machine learning model to built using this dataset? These are few of many questions that we want to find answers to in this project.

# Handling Missing Values

* The county column/feature have 100% missing values and there need to be dropped

* Fill all categorical features using the mode except the "image_url" and "description" which were filled using "ffill" method

* All float data types were filled using the mean values.

* <b> It is worthy of note that for the case of modelling, this is note the best approach to handle missing values. The best way is to group by fuel feature and do the imputation.

```python
df['Unnamed: 0'].nunique()
```

```python
df.posting_date.head()
```

```python
#Since size have 70% missing values, it will be wise we drop it
df = df.drop("size", axis = 1).reset_index(drop = True)
```

```python
df.shape
```

```python
df_obj.isnull().sum()
```

```python
#Checking for duplicates entries
dp = df.duplicated()
print(f"The number of duplicated row and column are: {df[dp].shape[0]} and {df[dp].shape[1]} respectively")
```

```python
df.manufacturer.fillna(value = 'ford', inplace = True)
df.model.fillna(value = 'f-150', inplace = True)
df.condition.fillna(value = 'excellent', inplace = True)
df.cylinders.fillna(value = '6 cylinders', inplace = True)
df.fuel.fillna(value = 'gas', inplace = True)
df.title_status.fillna(value = 'clean', inplace = True)
df.transmission.fillna(value = 'automatic', inplace = True)
df.VIN.fillna(value = '1HTMRAAM19J040080', inplace = True)
df.drive.fillna(value = '4wd', inplace = True)
df.type.fillna(value = 'sedan', inplace = True)
df.paint_color.fillna(value = 'white', inplace = True)
df.image_url.fillna(method = 'ffill', inplace = True)
df.description.fillna(method = 'ffill', inplace = True)
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].dtypes == 'float':
        df[col].fillna(df[col].mean(), inplace = True)
pass
df.isnull().sum()
```

# Preprocessing 

* The preprocessing phase is an import phase in this project. To build a good predictive model, the dataset must be processed to be in the form that our ML algorithm will understand. 

* The lat and long features are two powerful feature in this dataset as they have the strength to care for region, state, region_url and other features that are geo dependent. 

* With lat and long features, we can confidently drop the region, state and county features and still get a good performing model.

* Some of the categorical features will be preprocessed either by label enconding, one hot encoding or creating dummy variables from them.

## manufacturer Preprocessing
* The used car manufacturer though affects our target variable price, but is not a big contributor. It will be label enconded

```python
label_encoder = LabelEncoder()
df.manufacturer = label_encoder.fit_transform(df.manufacturer)
df.manufacturer.unique()
```

## model preprocessing
* The model of the car may affect price but is not an important feature to modelling the dataset.

```python
df.model = label_encoder.fit_transform(df.model)
df.model.nunique()
```

```python
df.model.value_counts().head(10)
```

## Condition preprocessing
* The condition feature will be converted to dummy variable/one hot encoded.

```python
df.condition = pd.get_dummies(df.condition)
df.condition.head()
```

## cylinders Preprocessing

```python
df.cylinders.unique()
```

```python
cylinder = {"old": ['8 cylinders', '4 cylinders', '6 cylinders', '10 cylinders',
                    '12 cylinders', '5 cylinders', '3 cylinders', 'other'],
            "new": [8,4,6,10,12,5,3,14]}
df.cylinders.replace(cylinder["old"], cylinder["new"], inplace = True)
df.cylinders.unique()
```
