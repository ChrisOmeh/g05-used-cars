```python
li = list(df.columns)
```

```python
input_li = ['price', 'year', 'manufacturer', 'model', 'condition', 'cylinders', 'fuel', 'odometer', 'title_status', 'transmission', 'drive', 'state', 'lat', 'long']
```

```python
for i,v in enumerate(li):
    if v not in input_li:
        df.drop(v, axis=1, inplace=True)
```

```python
df.head()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.info()
```

```python
df.isnull().value_counts()
```

```python
li1 = list(df['manufacturer'].unique())
li2 = list(df['model'].unique())
li3 = list(df['cylinders'].unique())
li4 = list(df['fuel'].unique())
li5 = list(df['title_status'].unique())
li6 = list(df['transmission'].unique())
li7 = list(df['drive'].unique())
li8 = list(df['state'].unique())
li9 = list(df['condition'].unique())
```

```python
def manu(text):
    for i, v in enumerate (li1):
        if text == v:
            return i
        
def mod(text):
    for i, v in enumerate (li2):
        if text == v:
            return i
        
def cyl(text):
    for i, v in enumerate (li3):
        if text == v:
            return i
        
def fuel(text):
    for i, v in enumerate (li4):
        if text == v:
            return i
        
def tit(text):
    for i, v in enumerate (li5):
        if text == v:
            return i
        
def tran(text):
    for i, v in enumerate (li6):
        if text == v:
            return i
        
def dri(text):
    for i, v in enumerate (li7):
        if text == v:
            return i
        
def sta(text):
    for i, v in enumerate (li8):
        if text == v:
            return i
        
def con(text):
    for i, v in enumerate (li9):
        if text == v:
            return i
```

```python
df['manufacturer'] = df['manufacturer'].apply(manu)
df['model'] = df['model'].apply(mod)
df['condition'] = df['condition'].apply(con)
df['cylinders'] = df['cylinders'].apply(cyl)
```

```python
df['fuel'] = df['fuel'].apply(fuel)
df['title_status'] = df['title_status'].apply(tit)
df['transmission'] = df['transmission'].apply(tran)
df['drive'] = df['drive'].apply(dri)
df['state'] = df['state'].apply(sta)
```
